```python
import requests as req
from bs4 import BeautifulSoup as bs

# PTT NBA 板
url = "https://www.ptt.cc/bbs/NBA/index.html"

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url)

# res.text 取得內文 指定 lxml 作為解析器
soup = bs(res.text, "lxml")

```

```python
# 第一個 <a></a>
soup.find('a')

# 全部 <a></a>，此時回傳 list
soup.find_all('a')

# 指定 list 某個元素的 html
soup.find_all('a')[2]
```

```python
# 取得 id 為 logo 的元素
a = soup.find(id = "logo")
# a   web element

# 取得所有 div，類別名稱為 r-ent，回傳為 list
divs = soup.find_all("div", class_="r-ent")
divs
```

```python
# 輸出 title
soup.select_one('title')

# 輸出 a
soup.select('a')

# 透過 class 名稱取得元素
soup.select("a.board")

# 透過 id 名稱取得元素
soup.select_one("#logo")

# 透過 attribute 取得元素
soup.select('a[class="board"]')
```

```python
# 取得單一節點的文字內容 (select_one 會回傳單一 bs element 物件，select 會回傳 list)

# .get_text() 只能針對單一元素取得內文
soup.select_one('title').get_text() #是字串
soup.select('a')[0].get_text()
```

```python
# 透過迭代取得所有 a 的文字內容
for a in soup.select('a'):
    # print(a.get_text())
    if a.has_attr('href'):
        print(a['href']) # a.get("href")
    else:
        print("="* 50)
        print(f"連結[{a.get_text()}]沒有 href 屬性")
        print("="* 50)
```
